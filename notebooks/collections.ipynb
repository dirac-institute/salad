```python
import lsst.daf.butler as dafButler
```

```python
butler = dafButler.Butler("../data")
```

```python
def n_exp(collection):
    return len(set(list(map(lambda x : x.id, list(
        butler.registry.queryDimensionRecords(
            "exposure", 
            datasets="differenceExp", 
            collections=collection
        )
    )))))

c = {}
for collection in butler.registry.queryCollections("DEEP/*"):
    c[collection] = n_exp(collection)
```

```python
" ".join({
    k: c[k] 
    for k in sorted(list(filter(lambda x : (c[x] > 30) and (len(x) == len("DEEP/20210516/A1d")), c)), key=lambda x : c[x], reverse=True)
}.keys())
```
