```python
from salad.serialize import read
import matplotlib.pyplot as plt
from astropy.timeseries import LombScargle
import astropy.units as u
from pathlib import Path
import astropy.table
import numpy as np

def load_summary(search_prefix, detector, snr, time_type, search_label, step="filtered"):
    # get recoveries at a certain step
    p = Path(f"/epyc/projects/salad/search/{search_prefix}/DEEP/20190403/A0c/detector_{detector}/snr_{snr}/{time_type}/{search_label}")
    return read(list(p.rglob(f"{step}_summary.pkl"))[0])
```

```python
# DEEP_20190403_A0c_detector_3_snr_3.0_cluster_2
matplotlib.rcParams.update({'font.size': 14})

s = load_summary("new", 3, 3.0, "regular", "asteroid")[2]
fig = summary_coadds_plot(s)
fig.tight_layout()
fig.savefig("bright.png")
plt.show()

cluster = load_summary("new", 3, 3.0, "regular", "asteroid")[2]
s = cluster.summary

fig = plt.figure(dpi=150, figsize=(5, 5))
axs = fig.subplots(3, 1, sharex=True, sharey=False)
# axs = np.atleast_2d(axs)
x = np.arange(len(s['expnum']))

plt.sca(axs[0])
plt.errorbar(
    x, 
    s['light_curve']['flux'], 
    yerr=s['light_curve']['sigma'],
    fmt='o',
    ms=2,
)
plt.ylabel("VR Flux")

plt.sca(axs[1])
plt.scatter(
    x, 
    s['light_curve']['mag'], 
    s=2,
)
plt.ylabel("VR Mag")

plt.sca(axs[2])
plt.scatter(
    x, 
    s['light_curve']['snr'], 
    s=2,
)
plt.ylabel("SNR")
plt.xlabel("Observation #")
fig.tight_layout()
plt.show()

from astropy.visualization import (ZScaleInterval, SqrtStretch,
                                   ImageNormalize, MinMaxInterval)

import matplotlib
matplotlib.rcParams.update({'font.size': 22})


v = s['image']
norm = ImageNormalize(
    v,
    interval=ZScaleInterval()
)

for i in range(10):
    snr = s['light_curve']['snr'][i]
    plt.imshow(v[i], norm=norm, cmap='gray_r')
    plt.title(f"SNR={snr:0.1f}")
    plt.axis("off")
    plt.show()
```
